```python
def same_first_last(nums):
  if len(nums)>=1:
    if nums[0]==nums[-1]:
      return True
    else:
      return False
  else:
    return False
print(same_first_last([1,2,3]))
```

```python
def date_fashion(you, date):
  if (date>=8 or you>=8) and not(date<=2 or you<=2):
    return 2
  elif date<=2 or you<=2:
    return 0
  else:
    return 1
print(date_fashion(5, 10))
```

```python
def in1to10(n, outside_mode):
  if 1<=n<=10 and outside_mode==False:
    return True
  elif (n<=1 or n>=10) and outside_mode==True:
    return True
  else:
    return False
print(in1to10(11, False))
```
